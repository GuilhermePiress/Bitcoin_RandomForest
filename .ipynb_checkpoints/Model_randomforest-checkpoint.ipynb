```python
btc_df=btc_total.copy()
btc_df=btc_df.dropna()
btc_df.reset_index(drop=True,inplace=True)
btc_df.drop(columns=['Lag_-1'],inplace=True)
btc_df.head()
```

```python
#The train_test will start with 80% of the entire dataframe
#The 80% is the line 388
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error

model_metrics_dataframe=pd.DataFrame(columns=['end_row_test','C_btc_true',
                                              'C_btc_predicted','RMSE_score','MAE','MAPE'])
end_row=int(btc_df.shape[0]*0.8)
for end_line in range(end_row+1,btc_df.shape[0]):
    
    X_train=btc_df.iloc[:end_line,1:]
    y_train=btc_df.iloc[:end_line,0]
    
    rf=RandomForestRegressor()
    rf.fit(X_train,y_train)
    
    X_test=btc_df.iloc[end_line,1:].values.reshape(1,-1)
    y_test=np.array([btc_df.iloc[end_line,0]])
    
    y_predicted=rf.predict(X_test)
    
    
    RMSE=mean_squared_error(y_test,y_predicted,squared=False)
    MAE=mean_absolute_error(y_test,y_predicted)
    MAPE=mean_absolute_percentage_error(y_test,y_predicted)
    
    model_metrics_dataframe=(model_metrics_dataframe
    .append({'end_row_test':end_line,'C_btc_true':y_test[0]
             ,'C_btc_predicted':y_predicted[0],'RMSE_score':RMSE,'MAE':MAE,'MAPE':MAPE},ignore_index=True))
```

```python
model_metrics_dataframe.describe()
```

```python
model_metrics_dataframe['Error_residue']=(model_metrics_dataframe['C_btc_true']-
                                          model_metrics_dataframe['C_btc_predicted'])
```

```python
from sklearn.metrics import r2_score
RMSE=mean_squared_error(model_metrics_dataframe['C_btc_true'],model_metrics_dataframe['C_btc_predicted'],squared=False)
MAE=mean_absolute_error(model_metrics_dataframe['C_btc_true'],model_metrics_dataframe['C_btc_predicted'])
MAPE=mean_absolute_percentage_error(model_metrics_dataframe['C_btc_true'],model_metrics_dataframe['C_btc_predicted'])
R2=r2_score(model_metrics_dataframe['C_btc_true'],model_metrics_dataframe['C_btc_predicted'])
```

```python
btc_final=btc_total.copy()
btc_final=btc_final.dropna()
btc_final.reset_index(drop=True,inplace=True)
btc_final['Diff']=btc_final['Lag_-1']-btc_final['C_btc']
btc_final['Label']=btc_final['Diff'].apply(lambda x: 1 if x>=0 else 0)
```

```python
btc_final[['C_btc','Lag_-1','Diff','Label']].head()
```

```python
#feature importance of the best model
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
X_train=btc_final.drop(columns=['C_btc','Lag_-1','Diff','Label'])
y_train=btc_final['Label']
    
rf.fit(X_train,y_train)

fig=plt.figure(figsize=(5,20))
plt.barh(btc_final.drop(columns=['C_btc','Lag_-1','Diff','Label']).columns,rf.feature_importances_)
```

```python
import shap
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_train)
shap.summary_plot(shap_values[0], X_train)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(btc_final.drop(columns=['C_btc','Lag_-1','Diff','Label'])
                                                    , btc_final['Label'], test_size=0.2, random_state=42)

sc=StandardScaler()

X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

param_grid={'max_depth' : np.arange(1,10)}
rf=RandomForestClassifier()

CV_rf=GridSearchCV(estimator=rf, param_grid=param_grid, cv=3)

CV_rf.fit(X_train,y_train)
y_predicted=CV_rf.predict(X_test)

print(classification_report(y_test,y_predicted),'Accuracy of train is: ',CV_rf.score(X_train,y_train))
```

```python
from sklearn.naive_bayes import BernoulliNB

BNB=BernoulliNB()

BNB.fit(X_train,y_train)
y_predicted=BNB.predict(X_test)

print(classification_report(y_test,y_predicted),'Accuracy of train is: ',BNB.score(X_train,y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

KNN=KNeighborsClassifier()
#Find the best K using Grid Search
GS_KNN=GridSearchCV(KNN,{'n_neighbors': np.arange(1, 50)}, cv=3)

GS_KNN.fit(X_train,y_train)

y_predicted=GS_KNN.predict(X_test)

print(classification_report(y_test,y_predicted),'Accuracy of train is: ',GS_KNN.score(X_train,y_train))
```

```python
from sklearn.svm import SVC

SVC=SVC()
SVC.fit(X_train,y_train)

y_predicted=SVC.predict(X_test)

print(classification_report(y_test,y_predicted),'Accuracy of train is: ',SVC.score(X_train,y_train))
```

```python
from xgboost import XGBClassifier
xgc=XGBClassifier()

xgc.fit(X_train,y_train)

y_predicted=xgc.predict(X_test)

print(classification_report(y_test,y_predicted),'Accuracy of train is: ',xgc.score(X_train,y_train))
```
